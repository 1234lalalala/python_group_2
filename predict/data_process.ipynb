```python
import pickle
df  = pickle.load(open('./pd_data.vec','rb'))
```

```python
df.head(1)
```

```python
def func(x):
    a = [i[1] for i in x]
    ret = {}
    for i in a:
        if i in ret:
            ret[i] += 1
        else:
            ret[i] = 1

    return max(list(ret.keys()),key = lambda x:ret[x])
# func(df['tags'][0])
df['tags'] = df['tags'].apply(func)
```

```python
train_x = df[['fans','ctime','duration','dimension','tname']]
train_y = df[['coin','danmaku','like','favorite','reply','view','share']]
```

```python
train_x['dimension0'] = train_x['dimension'].apply(lambda x : x[0])
train_x['dimension1'] = train_x['dimension'].apply(lambda x : x[1])
train_x['dimension2'] = train_x['dimension'].apply(lambda x : x[2])
```

```python
del train_x['dimension']
```

```python
import time
def time_hour(t):
    s  = time.localtime(t)
    dt = time.strftime("%H",s)
    return int(dt)
train_x['ctime'] = train_x['ctime'].apply(time_hour)
```

```python
train_x.head()
```

```python
import pandas as pd
train_x = pd.concat([train_x,pd.get_dummies(train_x['tname'])],axis=1)
```

```python
del train_x['tname']
```

```python
train_x.corr()
```

```python
y = train_y['view']
import sklearn
x_train = train_x.loc[0:2000]
x_test = train_x.loc[2000:]
y_train = y.loc[0:2000]
y_test = y.loc[2000:]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
predictions = model.predict(x_test)

model.score(x_test, y_test)
```
